```python
df100k['text_length'] = df100k['text'].apply(len)
sorted_df = df100k.sort_values(by='token_count', ascending=False)
sorted_df = sorted_df[sorted_df["text_length"] > 10000]
```

```python
# Filter the DataFrame to find entries where token_count is more than 8000
high_token_count_df = df100k[df100k['token_count'] > 2048]

# Find the minimum text_length from the filtered DataFrame
min_text_length = high_token_count_df['text_length'].min()

# Print the result
print("The smallest text_length where token_count is more than 8192 is:", min_text_length)
```
